```python
import xml.etree.ElementTree as et
import pandas as pd
#import numpy as np
import requests
```

```python
df = pd.DataFrame() 
```

```python
lista = [4751, 4752, 4755, 4756]

elementoslista = len(lista)
```

```python
dftemp = pd.DataFrame() 
contar = 0
posicao = 0
idlocal = str(lista[posicao])
elementoslista = len(lista)

while posicao < elementoslista:      
    idlocal = str(lista[posicao])
    dftemp = pd.DataFrame()
    url = 'http://servicos.cptec.inpe.br/XML/cidade/' + str(idlocal) + '/previsao.xml'
    header = { 'Accept': 'application/xml' }
    r = requests.get(url, headers=header) 
    tree =  et.ElementTree(et.fromstring(r.content))
    root=tree.getroot()

    dia = []
    tempo = []
    maxima=[]
    minima=[]
    iuv=[]
    cidade = []
    estado = []
    atualizacao = [] 

    for l in root.iter('dia'):
        dia.append(l.text)     
    for t in root.iter('tempo'):
        tempo.append(t.text) 
    for mx in root.iter('maxima'):
        maxima.append(mx.text) 
    for mn in root.iter('minima'):
        minima.append(mn.text) 
    for iu in root.iter('iuv'):
        iuv.append(iu.text) 

    for ci in root.iter('nome'):
        cidade = ci.text 
    for uf in root.iter('uf'):
        estado = uf.text 
    for atu in root.iter('atualizacao'):
        atualizacao = atu.text  

    dftemp = pd.DataFrame({'data_previsao':dia,'minima':minima ,'maxima':maxima, 'iuv':iuv})   

    dftemp['cidade'] = cidade
    dftemp['uf'] = estado
    dftemp['atualizacao'] = atualizacao
    
    contar = contar +1
    posicao = posicao+1
    
    print(idlocal)
    df = pd.concat([df,dftemp], ignore_index=True)
```
